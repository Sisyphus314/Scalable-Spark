```python
import csv
import dateutil
```

```python
def citibike_age(filename):
    with open(filename,'r') as fi:
        reader = csv.DictReader(fi)
        for row in reader:
            if row['usertype'] == 'Subscriber':
                age = 2020 - int(row["birth_year"])
                yield age
```

```python
count = {}
for age in citibike_age("citibike.csv"):
    count[age] = count.get(age, 0) + 1
```
